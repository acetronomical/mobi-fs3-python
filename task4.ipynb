**<a id="#">Task 4.1.2.</a>** Implement horizontal and vertical 3&times;3 Prewitt filters:

```python
def prewitt_h(img):
    img_copy = img.copy()
    for p in numpy.ndindex(img.shape):
        neighbourhood_sum = 0
        for q in numpy.ndindex(3, 3):
            i,j = p[0] + q[0] -1, p[1] + q[1] -1
            if 0 <= i <img.shape[0] and 0 <=j<img.shape[1]:
                neighbourhood_sum+=img[i,j] * (q[0] -1) * 1/6
        img_copy[p] = neighbourhood_sum
    return img_copy

    
def prewitt_v(img):
    img_copy = img.copy()
    for x,y in numpy.ndindex(img.shape):
        neighbourhood_sum = 0
        for f_x_v, f_y_v in numpy.ndindex(3, 3):
            f_x = f_x_v -1
            f_y = f_y_v + 1
            absolut_x = f_x + x
            absolut_y = f_y + y
            if 0 <= absolut_x <img.shape[0] and 0 <=absolut_y<img.shape[1]:
                neighbourhood_sum+=img[absolut_x,absolut_y] * (f_x) * 1/6
        img_copy[x,y] = neighbourhood_sum
    return img_copy
```

**<a id="#">Task 4.1.3.</a>** Test your above solution using the functions `prewitt_h` and `prewitt_v` for the previously loaded image and show the result:

```python
img_h = prewitt_h(img)
plt.figure(figsize=(15,8))
plt.imshow(img_h)

img_v = prewitt_v(img)
plt.figure(figsize=(15,8))
plt.imshow(img_v)
```

## Edge detection

**<a id="#">Task 4.2.1.</a>** Use your results from **Task 4.1.2 â€“ 4.1.3** to compute the *gradient magnitude* of the image:

```python
def prewitt_h(img):
    img_copy = img.copy()
    for p in numpy.ndindex(img.shape):
        neighbourhood_sum = 0
        for q in numpy.ndindex(3, 3):
            i,j = p[0] + q[0] -1, p[1] + q[1] -1
            if 0 <= i <img.shape[0] and 0 <=j<img.shape[1]:
                neighbourhood_sum+=img[i,j] * (q[0] -1) * 1/6
        img_copy[p] = numpy.sqrt(neighbourhood_sum)
    return img_copy

img_h = prewitt_h(img)
plt.figure(figsize=(15,8))
plt.imshow(img_h)
```

## Linear filtering by convolution (mean filter)

**<a id="#">Task 3.1.1.</a>** Load the image `data/astronaut.png` and show it:

**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape):
        neighbourhood_sum = 0
        count = 1
        for q in numpy.ndindex(size, size):
            i,j = p[0] + q[0] -n, p[1] + q[1] -n
            if 0 <= i <img.shape[0] and 0 <=j<img.shape[1]:
                neighbourhood_sum+=img[i,j]
                count +=1
            result[p] = neighbourhood_sum/count
    return result
```
